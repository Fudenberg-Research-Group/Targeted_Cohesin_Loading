```python
genome_length = 3*2.7 # Gbp
genome_len =3*2725521370 # The factor of 3 is considering 3 copies of each chromosome
experimental = 0.016764/3 # The read per bp measured experimentally
replica = 10
tot_motifs = 55073
occup_mean = 0.33182353 #Average occupancy from Sonmezer measurements
ctcf_tot_number = 217000
CTCFRate = 0.308 #The rate of CTCF at location of their motif sites, measured experimentally
nmultiprocess = 5
bound_fraction = 0.5 # fraction of Rad21 bound to chromatin
occup_mean_p = (ctcf_tot_number/3)*bound_fraction/tot_motifs
multiplication = occup_mean_p/occup_mean
ctcf_per_10_meg = (10*1e6 / (genome_length *1e9) * bound_fraction * ctcf_tot_number)# / occup_mean
ctcf_per_10_meg
#REGION = "chr1:30000000-37500000" imported from config
chrom, start, end = bioframe.parse_region_string(REGION)
size_bp = end - start
size_mb = size_bp / 1e6
simlen= size_bp
```

```python
def normalizedata(file, nmultiprocess= 5, replication = 10):
    
    lefs = h5py.File(file, 'r')['positions']
    lef = np.array(lefs)
    rad21reads = lef.shape[0]*lef.shape[1]*lef.shape[2]*nmultiprocess # calculating total number of extruders in simulation, 5 is for multiprocess 
    rad21persimlen = rad21reads/(replication*simlen) #factor of 10 is for 10 replication in simulations
    rad21normalizer = experimental/rad21persimlen
    ctcf_right = np.array(h5py.File(file, 'r')['CTCF_positions_right'])
    ctcf_left = np.array(h5py.File(file, 'r')['CTCF_positions_left'])
    ctcf_reads = (np.sum(np.array(ctcf_right))+ np.sum(np.array(ctcf_left)))*nmultiprocess
    ctcfpersimlen = ctcf_reads/simlen
    ctcfnormalizer = experimental/(ctcfpersimlen)
    return rad21normalizer, ctcfnormalizer
```

```python
# ---------- Utilities ----------
def pos_xy(df):
    x = df['ctcf'].to_numpy()
    y = df['Rad21'].to_numpy()
    m = (x > 0) & (y > 0) & np.isfinite(x) & np.isfinite(y)
    return x[m], y[m]

def kde_z(x, y):
    if len(x) == 0:
        return np.array([])
    xy = np.vstack([x, y])
    return gaussian_kde(xy)(xy)
```

```python
file1 = pd.read_csv('data/Rad21_vs_ctcf_reads_boundtime_random_configuration_main_figure.csv')
file2 = pd.read_csv('data/Rad21_vs_ctcf_reads_unbound_random_configuration.csv')
file3 = pd.read_csv('data/Rad21_vs_ctcf_reads_boundtime_regular_configuration.csv')
file4 = pd.read_csv('data/Rad21_vs_ctcf_reads_constoccup_random_configuration.csv')
file5 = pd.read_csv('data/Rad21_vs_ctcf_reads_boundtime_regular_configurationseparated.csv')  
#file5['ctcf']=file5['ctcfread']

for df in (file1, file2, file3, file4, file5):
    df['Rad21'] = df['Rad21']


# ---------- Positive subsets ----------
x1, y1 = pos_xy(file1)
x2, y2 = pos_xy(file2)
x3, y3 = pos_xy(file3)
x4, y4 = pos_xy(file4)
x5, y5 = pos_xy(file5) 

# ---------- Densities for color ----------
z1 = kde_z(x1, y1)
z2 = kde_z(x2, y2)
z3 = kde_z(x3, y3)
z4 = kde_z(x4, y4)
z5 = kde_z(x5, y5)    

# ---------- Box-plot data (five groups) ----------
g1 = pd.DataFrame({'Rad21': y1, 'Group': 'Variation on bound time'})
g2 = pd.DataFrame({'Rad21': y2, 'Group': 'Fixed bound time'})
g3 = pd.DataFrame({'Rad21': y3, 'Group': 'Regular CTCF arrangement,\nvariation on bound'})
g4 = pd.DataFrame({'Rad21': y4, 'Group': 'Const. occupancy'})
g5 = pd.DataFrame({'Rad21': y5, 'Group': 'Regular CTCF arrangement,\nSE fixed'})  # NEW
combined_df = pd.concat([g1, g2, g3, g4, g5], ignore_index=True)


# ---------- Positive subsets ----------
x1, y1 = pos_xy(file1)
x2, y2 = pos_xy(file2)
x3, y3 = pos_xy(file3)
x4, y4 = pos_xy(file4)
x5, y5 = pos_xy(file5) 

# ---------- Densities for color ----------
z1 = kde_z(x1, y1)
z2 = kde_z(x2, y2)
z3 = kde_z(x3, y3)
z4 = kde_z(x4, y4)
z5 = kde_z(x5, y5)  

# ---------- Box-plot data (five groups) ----------
g1 = pd.DataFrame({'Rad21': y1, 'Group': 'Variation on bound time'})
g2 = pd.DataFrame({'Rad21': y2, 'Group': 'Fixed bound time'})
g3 = pd.DataFrame({'Rad21': y3, 'Group': 'Regular CTCF arrangement,\nvariation on bound'})
g4 = pd.DataFrame({'Rad21': y4, 'Group': 'Const. occupancy'})
g5 = pd.DataFrame({'Rad21': y5, 'Group': 'Lattice+SE fixed'}) 
combined_df = pd.concat([g1, g2, g3, g4, g5], ignore_index=True)

# Fixed CTCF read value from file4 (if multiple, take median)
ctcf_fixed_value = np.median(x4) if len(x4) > 0 else np.nan

# ---------- Fonts ----------
axis_label_size = 18
tick_label_size = 14
colorbar_tick_size = 16
colorbar_label_size = 18
title_size = 18

# ---------- Figure with 2 rows x 4 columns ----------
fig, axes = plt.subplots(2, 4, figsize=(32, 12))
#fig.suptitle('CTCF spacing ≈ 200 kb', fontsize=18, y=0.98)

# Common axis limits (adjust if needed)
xlim = (40, 3115)
ylim = (0.9922, 8100)
norm = plt.matplotlib.colors.LogNorm()

def do_scatter(ax, x, y, z, title):
    sc = ax.scatter(x, y, c=z, cmap='viridis', norm=norm, s=100, alpha=0.8)
    ax.set_xscale('log'); ax.set_yscale('log')
    ax.set_xlabel('CTCF Reads', fontsize=axis_label_size)
    ax.set_ylabel('Rad21 Reads', fontsize=axis_label_size)
    ax.set_title(title, fontsize=title_size)
    ax.set_xlim(*xlim); ax.set_ylim(*ylim)
    cbar = fig.colorbar(sc, ax=ax, label='Density (log scale)')
    cbar.ax.tick_params(labelsize=colorbar_tick_size)
    cbar.set_label('Density (log scale)', fontsize=colorbar_label_size)
    ax.tick_params(axis='both', labelsize=tick_label_size)

# Top row: 4 scatters
do_scatter(axes[0, 0], x1, y1, z1, 'Variation on Bound Time')
do_scatter(axes[0, 1], x2, y2, z2, 'Fixed Bound Time')
do_scatter(axes[0, 2], x3, y3, z3, 'Regular CTCF arrangement, sep~35kb\nvariation on Bound')
do_scatter(axes[0, 3], x4, y4, z4, 'Const. Occupancy' if not np.isfinite(ctcf_fixed_value)
           else f'Const. Occupancy  |  Fixed CTCF ≈ {ctcf_fixed_value:.0f}')

# Bottom-left: scatter for file5
do_scatter(axes[1, 0], x5, y5, z5, 'Regular, sep ~ 200 kb fixed')

# Bottom-mid: Box plot comparing all five groups near fixed CTCF (from file4)
ax_box = axes[1, 1]
ax_box.clear()
target = np.median(x4) if len(x4) > 0 else np.nan
tol = 0.10

def in_band(x, target, tol):
    if not np.isfinite(target):
        return np.ones_like(x, dtype=bool)
    return (x >= target * (1 - tol)) & (x <= target * (1 + tol))

y1_band = y1[in_band(x1, target, tol)]
y2_band = y2[in_band(x2, target, tol)]
y3_band = y3[in_band(x3, target, tol)]
y4_band = y4[in_band(x4, target, tol)]
y5_band = y5[in_band(x5, target, tol)]

combined_df_fixed = pd.concat([
    pd.DataFrame({'Rad21': y1_band, 'Group': 'Variation on bound time'}),
    pd.DataFrame({'Rad21': y2_band, 'Group': 'Fixed bound time'}),
    pd.DataFrame({'Rad21': y3_band, 'Group': 'Regular CTCF arrangement,\nvariation on bound'}),
    pd.DataFrame({'Rad21': y4_band, 'Group': 'Const. occupancy'}),
    pd.DataFrame({'Rad21': y5_band, 'Group': 'Regular, sep ~ 200kb '}),
], ignore_index=True)

sns.boxplot(
    data=combined_df_fixed, x='Group', y='Rad21', ax=ax_box,
    showmeans=True, showfliers=False,
    meanprops={"marker": "o", "markerfacecolor": "white",
               "markeredgecolor": "black", "markersize": 8}
)
ax_box.set_yscale('log')
ax_box.set_xlabel('', fontsize=axis_label_size)
ax_box.set_ylabel('Rad21 (log scale)', fontsize=axis_label_size)
title_txt = 'Rad21 at fixed CTCF'
if np.isfinite(target):
    title_txt += f' ≈ {target:.0f} (±{int(tol*100)}%)'
ax_box.set_title(title_txt, fontsize=title_size)
ax_box.tick_params(axis='both', labelsize=tick_label_size)
for label in ax_box.get_xticklabels():
    label.set_rotation(15)
    label.set_ha('right')

ax_vio = axes[1, 2]
groups = [
    'Variation on bound time',
    'Fixed bound time',
    'Regular CTCF arrangement,\nvariation on bound',
    'Const. occupancy',
    'Regular, sep ~ 200 kb fixed'
]
vals_y = [y1, y2, y3, y4, y5]
vals_x = [x1, x2, x3, x4, x5]

def widths_from_x(xs, min_w=0.4, max_w=1.0):
    logs = [np.log10(a[a > 0]) for a in xs]
    iqrs = [np.subtract(*np.percentile(l, [75, 25])) if l.size else 0.0 for l in logs]
    max_iqr = max(iqrs) if any(iqrs) else 1.0
    return [(min_w + (iqr/max_iqr)*(max_w - min_w)) if max_iqr > 0 else min_w for iqr in iqrs]

positions = np.arange(1, len(groups) + 1)
widths = widths_from_x(vals_x, min_w=0.4, max_w=1.0)

vp = ax_vio.violinplot(
    vals_y,
    positions=positions,
    widths=widths,
    showmeans=False,
    showextrema=False,
    showmedians=False
)
for b in vp['bodies']:
    b.set_alpha(0.75)

for i, v in enumerate(vals_y, start=1):
    v = np.asarray(v); v = v[(v > 0) & np.isfinite(v)]
    if v.size:
        q1, med, q3 = np.percentile(v, [25, 50, 75])
        ax_vio.vlines(i, q1, q3, linewidth=3, colors='black', zorder=2)
        ax_vio.scatter([i], [med], s=35, facecolors='white', edgecolors='black', zorder=3)

ax_vio.set_yscale('log')
ax_vio.set_ylim(*ylim)
ax_vio.set_xticks(positions)
ax_vio.set_xticklabels(groups, rotation=15, ha='right')
ax_vio.set_xlabel('', fontsize=axis_label_size)
ax_vio.set_ylabel('Rad21 (log scale)', fontsize=axis_label_size)
ax_vio.set_title('Rad21 violins', fontsize=title_size)
ax_vio.tick_params(axis='both', labelsize=tick_label_size)

# Bottom-right: simple text panel (optional)
ax_txt = axes[1, 3]
ax_txt.axis('off')
plt.tight_layout(rect=[0, 0, 1, 0.97])
#plt.savefig('Figures/summary_with_box.pdf', dpi=500)
plt.show()
```

```python
for i in range(20):
    directory = './simulations/sims_random_bounda_%s/'%i
    path_dict = {}
    for name  in glob.glob(directory+'/folder_*'):
        path_dict[name.split('/sims_random_bounda_%s/'%i)[1][:]]= name
    path_dict = dict(sorted(path_dict.items()))
    filename = 'Rad21_vs_ctcf_reads_random_mod_bounda_chr1:30000000-37500000_run%s_withnear.csv'%i
    rfile = pd.read_csv('data/'+filename)
    rfile['ctcf'] = rfile['ctcf']*ctcfnormalizer*10*0.308#10 is for replica, as file['ctcf'] calculate reads in the first replica and 0.308 is for frip of 30.8 %
    rfile['Rad21'] = rfile['Rad21']*rad21normalizer
    rfile['ratio'] = rfile['Rad21']/rfile['ctcf']
    for name in list(path_dict.keys())[:]:
        params=[ast.literal_eval(i) for i in name.split('folder_')[1].split('_')[1::2]]
        face, back, clife, cof, life, slife, birth, pause, sep, site, replica, vel, steps= params
        c+=1
        mapN=mon*site*1
        directoryfile = directory+name+'/simulation_1'+'/LEFPositions.h5'
        lefs = h5py.File(directoryfile,'r')["positions"]
        lef_lefts = lefs[ba_time:,:,0].flatten()
        lef_rights = lefs[ba_time:,:,1].flatten()
        ctcfrightlist = np.array(h5py.File(directoryfile,'r')['CTCF_sites_right'])
        ctcfleftlist = np.array(h5py.File(directoryfile,'r')['CTCF_sites_left'])
        ctcfrightoccup = np.array(h5py.File(directoryfile,'r')['CTCF_positions_right'])
        lst = np.array(list(ctcfrightlist))# + list(ctcfleftlist))
        lef_positions = np.hstack((lef_lefts,lef_rights))
        for j, rows in rfile.iterrows():
            if rows['ctcfloc'] in ctcfrightlist:
                rfile.at[j, 'strand']='right'
            elif rows['ctcfloc'] in ctcfleftlist:
                rfile.at[j, 'strand']='left'
    rfile.to_csv('data/data_with_strand/%s'%filename)

    # 1) sort by ctcfloc so "next row" is meaningful
    r = rfile.sort_values('ctcfloc').reset_index(drop=False)  # keep old index if you want it
    
    # 2) next-row helpers
    r['strand_next']   = r['strand'].shift(-1)
    r['ctcfloc_next']  = r['ctcfloc'].shift(-1)
    r['Rad21_next']    = r['Rad21'].shift(-1)
    r['ctcf_next']     = r['ctcf'].shift(-1)
    
    # 3) same-strand mask
    same_strand = r['strand'] == r['strand_next']
    
    # 4) compute outputs for pairs where the next row has the same strand
    out = r.loc[same_strand, ['index','Rad21','ctcf','ctcfloc','strand',
                              'Rad21_next','ctcf_next','ctcfloc_next','strand_next']].copy()
    

    out['deltactcfloc'] = out['ctcfloc_next'] - out['ctcfloc']
    
    out['ratio_first']  = np.where(out['ctcf']     != 0, out['Rad21']     / out['ctcf'],     np.nan)
    out['ratio_second'] = np.where(out['ctcf_next']!= 0, out['Rad21_next']/ out['ctcf_next'], np.nan)
    
    out = out.rename(columns={'index':'row_index_first'})
    out = out[['row_index_first','strand','ctcfloc','ctcfloc_next','deltactcfloc',
               'Rad21','ctcf','ratio_first','Rad21_next','ctcf_next','ratio_second']]
    
    out_path = 'data/paired_same_strand_ctcfloc_deltas_%s.csv'%i
    out.to_csv(out_path, index=False)
```
